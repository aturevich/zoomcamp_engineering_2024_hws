# 1. Use a generator

Remember the concept of generator? Let's practice using them to futher our understanding of how they work.

Let's define a generator and then run it as practice.

**Answer the following questions:**

- **Question 1: What is the sum of the outputs of the generator for limit = 5?**
- **Question 2: What is the 13th number yielded**

I suggest practicing these questions without GPT as the purpose is to further your learning.

```python
def square_root_generator(limit):
    n = 1
    while n <= limit:
        yield n ** 0.5
        n += 1

# Example usage:
limit = 5
generator = square_root_generator(limit)

for sqrt_value in generator:
    print(sqrt_value)
```

```python
limit = 5
generator = square_root_generator(limit)

gen_list = []

for sqrt_value in generator:
    # print(sqrt_value)
    gen_list.append(sqrt_value)

print(sum(gen_list))
```

```python
limit = 13
generator = square_root_generator(limit)

for sqrt_value in generator:
    print(sqrt_value)
```

# 2. Append a generator to a table with existing data


Below you have 2 generators. You will be tasked to load them to duckdb and answer some questions from the data

1. Load the first generator and calculate the sum of ages of all people. Make sure to only load it once.
2. Append the second generator to the same table as the first.
3. **After correctly appending the data, calculate the sum of all ages of people.**




```python
def people_1():
    for i in range(1, 6):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 25 + i, "City": "City_A"}

for person in people_1():
    print(person)


def people_2():
    for i in range(3, 9):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 30 + i, "City": "City_B", "Occupation": f"Job_{i}"}


for person in people_2():
    print(person)
```

Re-use the generators from Exercise 2.

Load your first generator first, and then load the second one with merge. Since they have overlapping IDs, some of the records from the first load should be replaced by the ones from the second load.

After loading, you should have a total of 8 records, and ID 3 should have age 33.

```python
import pandas as pd
import duckdb
```

```python
df1 = pd.DataFrame(people_1())
df2 = pd.DataFrame(people_2())

conn = duckdb.connect(database=':memory:', read_only=False)

conn.execute("CREATE TABLE people AS SELECT * FROM df1")

sum_ages_first = conn.execute("SELECT SUM(Age) FROM people").fetchall()[0][0]
print(f"Sum of ages from the first generator: {sum_ages_first}")

conn.execute("ALTER TABLE people ADD COLUMN Occupation VARCHAR")

conn.register('df2', df2)
conn.execute("INSERT INTO people SELECT * FROM df2")

sum_ages_after_append = conn.execute("SELECT SUM(Age) FROM people").fetchall()[0][0]
print(f"Sum of all ages after appending the data: {sum_ages_after_append}")
```

```python
df1 = pd.DataFrame(people_1())
df2 = pd.DataFrame(people_2())


conn = duckdb.connect(database=':memory:', read_only=False)

conn.execute("""
CREATE TABLE people_new (
    ID INTEGER PRIMARY KEY,
    Name VARCHAR,
    Age INTEGER,
    City VARCHAR
)
""")

conn.execute("INSERT INTO people_new SELECT * FROM df1")
conn.execute("ALTER TABLE people_new ADD COLUMN Occupation VARCHAR")


conn.register('df2', df2)
conn.execute("""
INSERT INTO people_new (ID, Name, Age, City, Occupation)
SELECT * FROM df2
ON CONFLICT (ID) DO UPDATE SET Name=EXCLUDED.Name, Age=EXCLUDED.Age, City=EXCLUDED.City, Occupation=EXCLUDED.Occupation
""")

result = conn.execute("SELECT * FROM people_new ORDER BY ID").fetchdf()
print(result)
```

```python

sum_ages_result = conn.execute("SELECT SUM(Age) AS TotalAge FROM people_new").fetchall()
total_age = sum_ages_result[0][0]

print(f"Sum of ages of all the people: {total_age}")
```

```python
def people_1():
    for i in range(1, 6):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 25 + i, "City": "City_A"}

def people_2():
    for i in range(3, 9):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 30 + i, "City": "City_B", "Occupation": f"Job_{i}"}

people_db = {}

for person in people_1():
    people_db[person["ID"]] = person

for person in people_2():
    people_db[person["ID"]] = person

total_age = sum(person["Age"] for person in people_db.values())

print(total_age)
```

```python
df1 = pd.DataFrame(people_1())
df2 = pd.DataFrame(people_2())


conn = duckdb.connect(database=':memory:', read_only=False)

conn.execute("""
CREATE TABLE people_new (
    ID INTEGER PRIMARY KEY,
    Name VARCHAR,
    Age INTEGER,
    City VARCHAR
)
""")

conn.execute("INSERT INTO people_new SELECT * FROM df1")
conn.execute("ALTER TABLE people_new ADD COLUMN Occupation VARCHAR")

conn.register('df2', df2)

# Step 2: Update existing records in `people_new` with matching IDs from `df2`
conn.execute("""
UPDATE people_new SET
Name = (SELECT Name FROM df2 WHERE df2.ID = people_new.ID),
Age = (SELECT Age FROM df2 WHERE df2.ID = people_new.ID),
City = (SELECT City FROM df2 WHERE df2.ID = people_new.ID),
Occupation = (SELECT Occupation FROM df2 WHERE df2.ID = people_new.ID)
WHERE EXISTS (SELECT 1 FROM df2 WHERE df2.ID = people_new.ID)
""")

# Step 3: Insert new records from `df2` that do not exist in `people_new`
conn.execute("""
INSERT INTO people_new (ID, Name, Age, City, Occupation)
SELECT df2.ID, df2.Name, df2.Age, df2.City, df2.Occupation FROM df2
WHERE df2.ID NOT IN (SELECT ID FROM people_new)
""")

# Verify the results
result = conn.execute("SELECT * FROM people_new ORDER BY ID").fetchdf()
print(result)
```

```python
sum_ages_result = conn.execute("SELECT SUM(Age) AS TotalAge FROM people_new").fetchall()
total_age = sum_ages_result[0][0]

print(f"Sum of ages of all the people: {total_age}")
```
